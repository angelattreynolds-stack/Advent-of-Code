```python
from collections import defaultdict, Counter
filepath = '/Users/angelareynolds/Downloads/day1input.txt'


with open(filepath, 'r') as file:
    dial = 50
    sol = 0
    for line in file:
        line.strip()
        direction = line[0]
        distance = int(line[1:])
        if direction == 'L':
            dial = (dial - distance + 100)%100 #%100 simulates a circular dial forcing it to loop back to 0 after 99
        else:
            dial = (dial + distance)%100
        if dial == 0:
            sol += 1
    print(sol)
```

```python
from collections import defaultdict, Counter
filepath = '/Users/angelareynolds/Downloads/day1input.txt'


with open(filepath, 'r') as file:
    dial = 50
    sol = 0
    for line in file:
        line.strip()
        direction = line[0]
        distance = int(line[1:])
        for _ in range(distance):
            if direction == 'L':
                dial = (dial - 1 + 100)%100 #since we are looping through the range of the distance, this new formula counts clicks that pass 0
            else:
                dial = (dial + 1)%100
            if dial == 0:
                sol += 1
    print(sol)
```
